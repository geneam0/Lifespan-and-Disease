## Section 2: Linear Regression
**Requirements:**
+ Split your cleaned dataset using sklearn into a training and test set
+ Fit a linear regression to your training set
+ Visualize the regressor using matplotlib
+ Report on the error rate for your test set

```python
# Use pandas to extract features and prediction sets from our dataframe
X0=health.iloc[:,0:7]
y0=health.iloc[:,7]
```

Nice and then we use sklearn to create a training set and test set. I will reserve 20% of the dataset for testing.

```python
# Use sklearn to split our dataset
X_train0, X_test0, y_train0, y_test0 = train_test_split(X0, y0, test_size = 0.2, random_state = 1)
```

Let's use that training set and fit a Linear Regression to it:

```python
# Use sklearn to create and Linear Regressor and fit our data to it
model = LinearRegression()
model.fit(X_train0, y_train0)
print(model.coef_)
print(model.intercept_)
```

The model coefficients are our weights and the intercept is our y-intercept in y=mx+b. To make sure we're on the right track let's try a couple made up sample inputs and see what our model returns:

```python
# experiment with the regressor
y_pred0 = model.predict(X_test0)
print(y_test0)
print(y_pred0)
```

So the values for y_test0 seem to match with our y_pred0 values. The y_test0 is our actual values whereas the y_pred0 are the values we get from our linear regression model. We all love data visualizations so let's create a plot of our linear regression and our data. I will have the training data be blue dots, the test data green dots, with a red regression line cutting through the data.

```python
# Use matplotlib (or other visualization lib) to create a scatter plot and the regression to create a regression line

fig, axs = plt.subplots(4,2, figsize=(20, 25), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .3, wspace=.1)
axs = axs.ravel()
for x in range(7):

    # Training Data
    axs[x].scatter(X_train0.iloc[:,x], y_train0, color = "blue")
    # Testing Data
    axs[x].scatter(X_test0.iloc[:,x], y_test0, color = "green")
    # Labels
    axs[x].set_ylabel("Life Span Outcome in Years")
    axs[x].set_xlabel(health.columns.values[x])

    # Plotting the trained regression line
    numbers = np.linspace(-1000,1000,1000)
    y = numbers*model.coef_[x] + model.intercept_
    axs[x].plot(numbers, y, color="red")
    axs[x].set_ylim(0,100)

    if(x==0):
        axs[x].set_xlim(-5,25)
    elif(x==1):
        axs[x].set_xlim(-5,200)
    elif(x==2):
        axs[x].set_xlim(0,250)
    elif(x==6):
        axs[x].set_xlim(0,25)
    else:
        axs[x].set_xlim(0,6)
```

```python
# Use linear regression model to report accuracy/error rate on the test set
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test0, y_pred0))
print('Mean Squared Error:', metrics.mean_squared_error(y_test0, y_pred0))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test0, y_pred0)))
print('R-Squared Score:', metrics.r2_score(y_test0, y_pred0))
```

I was shocked upon seeing the $r^2$ score. I didn't even think it was possible. I looked online and saw a couple of explanations. One on StackOverflow explains that the equation for r^2:
$$1 - \frac{(\text{residual sum of squares})}{(\text{total sum of squares})}$$
makes it so that if your denominator is small, you get a high negative number. Another explained that a low, negative value means that I would have gotten a better predictive score if I just used the mean each time. Still, I plugged both y_test_actual and y_predicted into an online $r^2$ calculator and got an $r^2$ value of $0.641$, which I don't think is too bad.

## Section 4: Classification with a Neural Network (or classifier of choice)
**Requirements:**

I will be using a neural network for the classifier model. First, let's split out our data into a training and test set. I will hold 20% of the data for testing.

```python
# Use sklearn to split our dataset
features = ['Age','EDUC','CDR']
x = disease[features]
y = disease['Group']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

Before we train we also need to scale our data for our neural network. This is because neural networks are suspetible to overfitting on features that have a higher magnitude. To scale this data I will use a [MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html)

```python
# Use sklearn to scale your feature vector (if you use a neural network)

scaler = MinMaxScaler()
scaler.fit_transform(X_train)
scaler.fit_transform(X_test)
```

Alright now that our data is ready, let's create a neural network. For this project I will use a neural network with 3 hidden layers. My neural network is a connected graph using the relu activation functions in the middle and sigmoid function at the last layer. The sigmoid function makes it easy for us to map to 0 or 1, whether or not the patient has Alzheimer's.

```python
# source code from https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/
# define the keras model - connected graph using relu and sigmoid activation functions
model = Sequential()
model.add(Dense(12, input_dim=3, activation='relu')) # input is 3
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # last layer being sigmoid allows us to easily map to either 0 or 1
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=100, batch_size=10)
```

```python
_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

NICE! let's print the weights.

```python
model.get_weights()
```

And try our model on the test set.

```python
# Error in testing data
predictions = model.predict(X_test)
# round predictions 
rounded = [round(x[0]) for x in predictions]
correct=0

for x in range(len(rounded)):
    if(rounded[x]==y_test.iloc[x]):
        correct+=1
error = 100*correct/len(rounded)
print(error)
```

Great, this means that we have an accuracy of 99.25% for our Alzheimer's neural network model on our training set and 100% on our test set. I am a bit surprised at how well it worked given our lack of feature variables. I would expect that we would need more to verify whether the patient has Alzheimer's, but I could be wrong and those three worked just as well.

```python
# Initialize a SVM Classifier with your choice of parameters

clf = svm.SVC(kernel="linear")
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
clf_accuracy = accuracy_score(y_test, y_pred)
print(clf_accuracy)
```
